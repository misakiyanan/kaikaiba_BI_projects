对于搜索广告是自带天然精准属性的，转化能力也是比较强的，主要原因是其受众标签是上下文搜索查询定向，反应了客户的强烈需求，而且客户当前目的性强，当前阶段的需求就不容易被打断，因此其受市场的重视也比较高，所占广告份额也是最大的，但是鉴于其比较贵，所以相对现在比较便宜的信息流广告来计算转化的时候还是要综合对比。

信息流广告主要在于信息流，就是长得像信息的广告，让人一看还以为是一条消息，然后点进去一看是广告，如果进来的受众有需求的话才有可能产生转化，所以信息流广告的随机性太强，所面对受众的需求随机性也比较强，这样会导致信息流广告效果大大降低，但是信息流广告有一个优点就是便宜，可以大量展现，大量投放，这在一定程度上弥补了其不精准性，还有其标签也是很不准的，而且每个公司对客户的标签都是不一样的，这里要强调一点信息流广告并不精准，广告公司所吹嘘的精准投放，N多标签，N多受众，是很不靠谱的，只能从一定程度上缓解信息流广告的不精准性，其核心还是靠大量的展现来进行转化。
